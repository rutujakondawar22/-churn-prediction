```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

df = pd.read_csv(r'C:\Users\Admin\churn-prediction\data\WA_Fn-UseC_-Telco-Customer-Churn.csv')
print("Shape:", df.shape)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print("Missing after conversion:", df['TotalCharges'].isnull().sum())
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
print("Missing after fix:", df['TotalCharges'].isnull().sum())
```

```python
df.drop('customerID', axis=1, inplace=True)
print("Columns remaining:", df.shape[1])
```

```python
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
print(df['Churn'].value_counts())
```

```python
binary_cols = ['gender', 'Partner', 'Dependents', 'PhoneService',
               'PaperlessBilling', 'MultipleLines']

le = LabelEncoder()
for col in binary_cols:
    df[col] = le.fit_transform(df[col])

print("Binary encoding done!")
df.head(2)
```

```python
df = pd.get_dummies(df, columns=[
    'InternetService', 'OnlineSecurity', 'OnlineBackup',
    'DeviceProtection', 'TechSupport', 'StreamingTV',
    'StreamingMovies', 'Contract', 'PaymentMethod'
])

print("Shape after encoding:", df.shape)
```

```python
df['revenue_per_tenure'] = df['TotalCharges'] / (df['tenure'] + 1)
print("New feature created!")
df[['TotalCharges', 'tenure', 'revenue_per_tenure']].head()
```

```python
scaler = StandardScaler()
df[['tenure', 'MonthlyCharges', 'TotalCharges', 'revenue_per_tenure']] = scaler.fit_transform(
    df[['tenure', 'MonthlyCharges', 'TotalCharges', 'revenue_per_tenure']]
)
print("Scaling done!")
df.head(2)
```

```python
df.to_csv('../data/cleaned_churn.csv', index=False)
print("Cleaned data saved! Shape:", df.shape)
```
